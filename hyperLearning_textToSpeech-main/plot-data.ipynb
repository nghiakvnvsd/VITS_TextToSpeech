```python
import matplotlib.pyplot as plt
```

```python
with open("../checkpoints/exp5/train.log", "r") as f:
    data = f.readlines()
```

```python
data = [line.strip() for line in data if "Step" in line]
data[:4]
```

```python
# steps
steps = [int(line.split("|")[1].strip().replace("Step ", "")) for line in data]

# loss
mel_loss = [float(line.split("|")[6].strip().replace("Mel ", "")) for line in data]
duration_loss = [float(line.split("|")[7].strip().replace("Dur ", "")) for line in data]
```

```python
plt.title("Mel loss")
plt.plot(mel_loss[-100000:])
plt.show()
```

```python
plt.title("Duration loss")
plt.plot(duration_loss[-1000:])
plt.axhline(y=0, c="red")
plt.show()
```
